```python
# Imports here
import pandas as pd
import matplotlib.pyplot as plt
```

## Function loads csv file into pandas dataframe and prints however many rows you want

```python
def loadAndPrint (filepath, rows):
    df = pd.read_csv(filepath)
    print(df.head(rows))
```

## Make horizontal bar chart showing cost of the top 10 most baught items in comparison

```python
def makeBarChart (filepath):
    df = pd.read_csv(filepath, nrows=25)

    plt.figure(figsize=(24, 12))
    
    plt.bar(df['product_name'], df['cost'], color='skyblue')
    
    plt.xlabel('Product Name')
    plt.ylabel('Price')
    plt.title('Comparison of Products by Price (Top 25)')
    
    plt.xticks(rotation=45, ha='right')

    plt.savefig("top_25_ranked_products.svg", format="svg")
    plt.show()
```

## Print the first 25 rows of the csv file

```python
path = 'datasets/abs080922.csv'
loadAndPrint(path, 25)
```

## Make bar chart showing the top 25 best ranked product, comparing them by price

```python
makeBarChart(path)
```
